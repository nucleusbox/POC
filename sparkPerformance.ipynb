```python
import pandas as pd
import numpy as np
```

```python
nrows = 10000000

ndata_columns = 100

# Generate some string categories
ncategories = 10

categories = [pd.util.testing.rands(10) for _ in range(ncategories)]
cats = pd.Series(categories).repeat(nrows / ncategories)

data = {
    'group': cats
}
column_names = ['group']

# Generate some double data
for i in range(ndata_columns):
    name = 'data{}'.format(i)
    data[name] = np.random.randn(nrows)
    column_names.append(name)

%time df = pd.DataFrame(data, columns=column_names)
df.head()
```

```python
df.to_csv('test_dataset_10M.csv', index=False, header=False)
```
